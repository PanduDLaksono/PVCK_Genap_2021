```python
# 2. Face Detection
cascade_wajah = cv.CascadeClassifier('/content/drive/MyDrive/Polinema/PCVK_2021/images/haarcascades/haarcascade_frontalface_alt.xml')
```

```python
# JOKOWI
jokowi = cv.imread('/content/drive/MyDrive/Polinema/PCVK_2021/images/facedet/jokowi.jpg')
jokowi_1 = cv.imread('/content/drive/MyDrive/Polinema/PCVK_2021/images/facedet/jokowi.jpg')
roi_wajah1 = cascade_wajah.detectMultiScale(jokowi_1) 

roi_wajah = cascade_wajah.detectMultiScale(jokowi) 
for(x,y,w,h) in roi_wajah: 
 cv.rectangle(jokowi,(x,y),(x+w,y+h),(0,255,0),3) 
plt.figure(figsize = (10,10)) 

f, axarr = plt. subplots(1,2,figsize=(15,5))
plt.subplot(121),plt.imshow(cv.cvtColor(jokowi_1, cv.COLOR_BGR2RGB))
axarr[0].imshow(cv.cvtColor(jokowi_1, cv.COLOR_BGR2RGB))
plt.subplot(122),plt.imshow(cv.cvtColor(jokowi, cv.COLOR_BGR2RGB))
axarr[1].imshow(cv.cvtColor(jokowi, cv.COLOR_BGR2RGB))
plt.show()
```
